```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 1: Calculo de velocidad final (Secuencial)
# =========================================================
# --- ENTRADAS ---
# Definimos las variables con los datos del enunciado
velocidad_inicial = 0.0  # en m/s
aceleracion = 0.8        # en m/s^2
tiempo = 30.0            # en segundos

# --- PROCESO ---
# Aplicamos la fórmula física del MRUV
velocidad_final = velocidad_inicial + (aceleracion * tiempo)

# --- SALIDA ---
# Mostramos el resultado en la pantalla
print("=== RESULTADO DEL CÁLCULO ===")
print(f"La velocidad final del automóvil es: {velocidad_final} m/s")
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 2: Promedio de Calificaciones (Secuencial)
# =========================================================

# --- ENTRADAS ---
# Solicitamos las 4 calificaciones de manera interactiva
nota1 = float(input("Ingrese la calificación del examen 1: "))
nota2 = float(input("Ingrese la calificación del examen 2: "))
nota3 = float(input("Ingrese la calificación del examen 3: "))
nota4 = float(input("Ingrese la calificación del examen 4: "))

# --- PROCESO ---
# Estructura matemática secuencial para obtener la media aritmética
suma_notas = nota1 + nota2 + nota3 + nota4
promedio_final = suma_notas / 4

# --- SALIDA ---
# Mostramos el resultado con estructura clara y dos decimales
print("\n" + "="*30)
print("       REPORTE DE NOTAS")
print("="*30)
print(f"-> Nota 1: {nota1}")
print(f"-> Nota 2: {nota2}")
print(f"-> Nota 3: {nota3}")
print(f"-> Nota 4: {nota4}")
print("-"*30)
print(f"PROMEDIO FINAL: {promedio_final:.2f}")
print("="*30)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 3: Porcentajes de Socios (Estructura Secuencial)
# =========================================================

# --- ENTRADAS ---
inv1 = float(input("Ingrese el monto invertido por la Persona 1 (S/): "))
inv2 = float(input("Ingrese el monto invertido por la Persona 2 (S/): "))
inv3 = float(input("Ingrese el monto invertido por la Persona 3 (S/): "))

# --- PROCESO ---
# 1. Calculamos el capital total acumulado
total_inversion = inv1 + inv2 + inv3

# 2. Calculamos los porcentajes individuales correspondientes
pct1 = (inv1 / total_inversion) * 100
pct2 = (inv2 / total_inversion) * 100
pct3 = (inv3 / total_inversion) * 100

# --- SALIDA ESTRUCTURADA ---
print("\n" + "=" * 53)
print("     REPORTE DE CONSTITUCIÓN DE EMPRESA")
print("=" * 53)
print(f"    Capital Social Total : S/ {total_inversion:.2f}")
print("-" * 53)
print(f"-> Socio 1: Aporte S/ {inv1:<8.2f} | Porcentaje: {pct1:.2f}%")
print(f"-> Socio 2: Aporte S/ {inv2:<8.2f} | Porcentaje: {pct2:.2f}%")
print(f"-> Socio 3: Aporte S/ {inv3:<8.2f} | Porcentaje: {pct3:.2f}%")
print("=" * 53)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 4: Sueldo y Comisiones (Estructura Secuencial)
# =========================================================

# --- ENTRADAS ---
sueldo_base = float(input("Ingrese el sueldo base del vendedor (S/): "))
venta1 = float(input("Ingrese el monto de la venta 1 (S/): "))
venta2 = float(input("Ingrese el monto de la venta 2 (S/): "))
venta3 = float(input("Ingrese el monto de la venta 3 (S/): "))

# --- PROCESO ---
# 1. Calculamos el acumulado de las tres ventas
total_ventas = venta1 + venta2 + venta3

# 2. Calculamos el 25% de comisión extra sobre el total vendido
comisiones = total_ventas * 0.25

# 3. Calculamos el dinero total que recibirá en el mes
sueldo_total = sueldo_base + comisiones

# --- SALIDA ESTRUCTURADA ---
print("\n" + "=" * 45)
print("         BOLETA DE PAGO MENSUAL")
print("=" * 45)
print(f"Sueldo Fijo Base     : S/ {sueldo_base:.2f}")
print(f"Total en Ventas (3)  : S/ {total_ventas:.2f}")
print("-" * 45)
print(f"-> GANANCIA POR COMISIÓN (25%): S/ {comisiones:.2f}")
print("-" * 45)
print(f"TOTAL NETO A RECIBIR          : S/ {sueldo_total:.2f}")
print("=" * 45)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 5: Rendimiento de Capital (Estructura Secuencial)
# =========================================================

# --- ENTRADAS ---
capital_inicial = float(input("Ingrese el capital que desea invertir (S/): "))

# --- PROCESO ---
# El banco paga 2% mensual (0.02) y la inversión dura 1 año (12 meses)
tasa_mensual = 0.02
meses_año = 12

# Cálculo de la ganancia obtenida tras un año
ganancia_final = capital_inicial * tasa_mensual * meses_año

# Cálculo del monto total (Capital + Ganancia)
monto_total = capital_inicial + ganancia_final

# --- SALIDA ESTRUCTURADA ---
print("\n" + "=" * 45)
print("     REPORTE DE RENDIMIENTO FINANCIERO ANUAL")
print("=" * 45)
print(f"Capital Invertido    : S/ {capital_inicial:.2f}")
print(f"Tasa de Interés      : 2.0% mensual (24% anual)")
print("-" * 45)
print(f"-> DINERO GANADO (1 AÑO) : S/ {ganancia_final:.2f}")
print("-" * 45)
print(f"MONTO TOTAL ACUMULADO    : S/ {monto_total:.2f}")
print("=" * 45)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 6: Tiempo Vivido Aproximado (Estructura Secuencial)
# =========================================================

# --- ENTRADAS ---
edad_años = int(input("Ingrese la edad de la persona en años: "))

# --- PROCESO ---
# Definimos las equivalencias estándar de un año calendario común
MESES_POR_AÑO = 12
SEMANAS_POR_AÑO = 52
DIAS_POR_AÑO = 365
HORAS_POR_AÑO = 365 * 24  # 8760 horas

# Cálculos basados en la edad ingresada (Corregido 'SEMANAS_POR_AÑO')
total_meses = edad_años * MESES_POR_AÑO
total_semanas = edad_años * SEMANAS_POR_AÑO
total_dias = edad_años * DIAS_POR_AÑO
total_horas = edad_años * HORAS_POR_AÑO

# --- SALIDA ESTRUCTURADA ---
print("\n" + "=" * 45)
print("         REPORTE DE TIEMPO VIVIDO")
print("=" * 45)
print(f"Edad analizada       : {edad_años} años")
print("-" * 45)
print(f"-> Meses vividos     : {total_meses:,} meses")
print(f"-> Semanas vistas    : {total_semanas:,} semanas")
print(f"-> Días vividos      : {total_dias:,} días")
print(f"-> Horas vividas     : {total_horas:,} horas")
print("-" * 45)
print("(*) Nota: Los cálculos son aproximaciones estándar.")
print("=" * 45)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 7: Cobro de Estacionamiento (Condicional con Validación)
# =========================================================

print("--- SISTEMA DE CONTROL DE ESTACIONAMIENTO ---")
print("Nota: Ingrese valores numéricos enteros (Ejemplo: 5 horas, 30 minutos)\n")

# --- ENTRADAS CON VALIDACIÓN BÁSICA ---
try:
    horas = int(input("Ingrese la cantidad de horas completas: "))

    # Explicamos al usuario que si es media hora, introduzca 30
    print("[Si fue media hora, ingrese 30. Si fue un cuarto de hora, ingrese 15]")
    minutos = int(input("Ingrese los minutos adicionales (0 a 59): "))

    # --- PROCESO ---
    TARIFA_HORA = 5.00
    TARIFA_MINUTO = 0.10

    # Calculamos el costo base por las horas completas
    costo_horas = horas * TARIFA_HORA

    # Validamos y evaluamos si hay minutos adicionales válidos
    if minutos > 0:
        costo_minutos = minutos * TARIFA_MINUTO
    else:
        costo_minutos = 0.0

    # Calculamos el monto final acumulado
    total_pagar = costo_horas + costo_minutos

    # --- SALIDA ESTRUCTURADA ---
    print("\n" + "=" * 45)
    print("         TICKET DE ESTACIONAMIENTO")
    print("=" * 45)
    print(f"Tiempo registrado    : {horas}h y {minutos}m")
    print(f"Costo por horas      : S/ {costo_horas:.2f}")
    print(f"Costo por minutos    : S/ {costo_minutos:.2f}")
    print("-" * 45)
    print(f"TOTAL NETO A PAGAR   : S/ {total_pagar:.2f}")
    print("=" * 45)

except ValueError:
    print("\n" + "!" * 45)
    print("ERROR: Por favor, ingrese solo números enteros.")
    print("Ejemplo: Si es media hora, ejecute de nuevo e ingrese 30.")
    print("!" * 45)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 8: Área de Terreno Irregular (Estructura Secuencial)
# =========================================================

print("--- SISTEMA DE CÁLCULO DE ÁREAS - CONSTRUCTORA ---")
print("Por favor, ingrese las medidas del terreno en metros (m).\n")

try:
    # --- ENTRADAS ---
    lado_A = float(input("Ingrese la medida del lado A (Altura izquierda): "))
    lado_B = float(input("Ingrese la medida del lado B (Base inferior): "))
    lado_C = float(input("Ingrese la medida del lado C (Altura derecha): "))

    # Validación lógica geométrica elemental
    if lado_A <= 0 or lado_B <= 0 or lado_C <= 0:
        print("\n[ERROR]: Las medidas deben ser mayores a cero.")
    elif lado_A < lado_C:
        print("\n[ERROR]: El lado A debe ser mayor o igual al lado C según el diseño.")
    else:
        # --- PROCESO ---
        # 1. Calcular el área de la sección rectangular inferior
        area_rectangulo = lado_B * lado_C

        # 2. Calcular el área de la sección triangular superior
        altura_triangulo = lado_A - lado_C
        area_triangulo = (lado_B * altura_triangulo) / 2

        # 3. Calcular el área total combinada
        area_total = area_rectangulo + area_triangulo

        # --- SALIDA ESTRUCTURADA ---
        print("\n" + "=" * 45)
        print("          REPORTE TÉCNICO DEL TERRENO")
        print("=" * 45)
        print(f"Medidas: Lado A = {lado_A}m | Lado B = {lado_B}m | Lado C = {lado_C}m")
        print("-" * 45)
        print(f"-> Área de la Base (Rectángulo) : {area_rectangulo:.2f} m²")
        print(f"-> Área del Techo (Triángulo)  : {area_triangulo:.2f} m²")
        print("-" * 45)
        print(f"ÁREA TOTAL DEL TERRENO         : {area_total:.2f} m²")
        print("=" * 45)

except ValueError:
    print("\n" + "!" * 45)
    print("ERROR: Ingrese únicamente valores numéricos válidos.")
    print("!" * 45)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 9: Pago con Descuento y Vuelto (Estructura Condicional)
# =========================================================

print("--- SISTEMA DE CAJA - CÁLCULO DE COMPRAS ---")

try:
    # --- ENTRADAS ---
    monto_base = float(input("Ingrese el monto base de la compra (S/): "))
    pct_descuento = float(input("Ingrese el porcentaje de descuento (%): "))
    efectivo_cliente = float(input("Ingrese el dinero en efectivo entregado (S/): "))

    # Validación de coherencia para valores negativos
    if monto_base < 0 or pct_descuento < 0 or efectivo_cliente < 0:
        print("\n[ERROR]: No se permiten valores negativos.")
    else:
        # --- PROCESO ---
        # 1. Calcular el descuento y el total final de la compra
        monto_descuento = monto_base * (pct_descuento / 100)
        total_pagar = monto_base - monto_descuento

        # 2. Estructura condicional para evaluar si el efectivo alcanza y calcular el cambio
        if efectivo_cliente >= total_pagar:
            cambio_vuelto = efectivo_cliente - total_pagar

            # --- SALIDA ESTRUCTURADA ---
            print("\n" + "=" * 45)
            print("         COMPROBANTE DE PAGO REPORTE")
            print("=" * 45)
            print(f"Monto Base Compra    : S/ {monto_base:.2f}")
            print(f"Descuento ({pct_descuento}%)     : S/ {monto_descuento:.2f}")
            print("-" * 45)
            print(f"TOTAL NETO A PAGAR   : S/ {total_pagar:.2f}")
            print(f"Efectivo Recibido    : S/ {efectivo_cliente:.2f}")
            print("-" * 45)
            print(f"CAMBIO / VUELTO      : S/ {cambio_vuelto:.2f}")
            print("=" * 45)
        else:
            dinero_faltante = total_pagar - efectivo_cliente
            print("\n" + "!" * 45)
            print("TRANSACCIÓN RECHAZADA")
            print(f"El efectivo (S/ {efectivo_cliente:.2f}) es insuficiente.")
            print(f"Monto total a pagar  : S/ {total_pagar:.2f}")
            print(f"Dinero faltante      : S/ {dinero_faltante:.2f}")
            print("!" * 45)

except ValueError:
    print("\n" + "!" * 45)
    print("ERROR: Por favor, ingrese montos numéricos válidos.")
    print("!" * 45)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 10: Desglose de Cajero Automático (Estructura Secuencial/Ciclo)
# =========================================================

print("--- SISTEMA DE DISPENSACIÓN DE EFECTIVO ---")

try:
    # --- ENTRADAS ---
    monto_retirar = int(input("Ingrese la cantidad de dinero a retirar (S/): "))

    # Validación para asegurar que el monto sea un valor positivo lógico
    if monto_retirar <= 0:
        print("\n[ERROR]: Ingrese un monto mayor a cero.")
    else:
        # Guardamos el monto original para el reporte final
        saldo_restante = monto_retirar

        # Definimos las denominaciones disponibles de mayor a menor
        denominaciones = [200, 100, 50, 20, 10, 5, 1]

        # --- SALIDA ESTRUCTURADA (Inicio) ---
        print("\n" + "=" * 45)
        print("         RECIBO DE RETIRO - CAJERO")
        print("=" * 45)
        print(f"Monto Solicitado     : S/ {monto_retirar}")
        print("-" * 45)
        print("DESGLOSE DE EFECTIVO ENTREGADO:")

        # --- PROCESO ---
        # Recorremos cada denominación aplicando divisiones y residuos
        for valor in denominaciones:
            cantidad = saldo_restante // valor
            saldo_restante = saldo_restante % valor

            # Solo mostramos las denominaciones que realmente se van a entregar
            if cantidad > 0:
                if valor >= 10:
                    print(f"-> {cantidad:<2} billete(s) de  : S/ {valor}")
                else:
                    print(f"-> {cantidad:<2} moneda(s) de   : S/ {valor}")

        print("-" * 45)
        print("Por favor, retire su dinero y su recibo.")
        print("=" * 45)

except ValueError:
    print("\n" + "!" * 45)
    print("ERROR: Ingrese únicamente montos enteros sin decimales.")
    print("!" * 45)
```

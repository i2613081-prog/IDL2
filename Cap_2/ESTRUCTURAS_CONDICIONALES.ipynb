```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 1: Compra de Llantas (Condicional Simple)
# =========================================================

# --- ENTRADAS ---
cantidad_llantas = int(input("Ingrese la cantidad de llantas a comprar: "))

# --- PROCESO (Estructura Condicional) ---
if cantidad_llantas < 5:
    precio_unitario = 150.0
else:
    precio_unitario = 120.0

# Cálculo del monto total
total_pagar = cantidad_llantas * precio_unitario

# --- SALIDA ESTRUCTURADA ---
print("\n" + "=" * 35)
print("        BOLETA DE VENTA - LLANTERA")
print("=" * 35)
print(f"Cantidad de llantas : {cantidad_llantas}")
print(f"Precio por unidad   : S/ {precio_unitario:.2f}")
print("-" * 35)
print(f"TOTAL A PAGAR       : S/ {total_pagar:.2f}")
print("=" * 35)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 2: Mayor y Menor de Tres Números (Condicional Anidado)
# =========================================================

# --- ENTRADAS ---
num1 = float(input("Ingrese el primer número: "))
num2 = float(input("Ingrese el segundo número: "))
num3 = float(input("Ingrese el tercer número: "))

# --- PROCESO PARA HALLAR EL MAYOR ---
if num1 >= num2 and num1 >= num3:
    mayor = num1
elif num2 >= num1 and num2 >= num3:
    mayor = num2
else:
    mayor = num3

# --- PROCESO PARA HALLAR EL MENOR ---
if num1 <= num2 and num1 <= num3:
    menor = num1
elif num2 <= num1 and num2 <= num3:
    menor = num2
else:
    menor = num3

# --- SALIDA ESTRUCTURADA ---
print("\n" + "=" * 40)
print("         RESULTADO DE LA COMPARACIÓN")
print("=" * 40)
print(f"Números evaluados: [{num1}], [{num2}], [{num3}]")
print("-" * 40)
print(f"-> EL NÚMERO MAYOR ES : {mayor}")
print(f"-> EL NÚMERO MENOR ES : {menor}")
print("=" * 40)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 3: Control de Tributación (Condicional Compuesta)
# =========================================================

print("--- SISTEMA DE EVALUACIÓN SUNAT / TRIBUTARIA ---")

try:
    # --- ENTRADAS ---
    edad = int(input("Ingrese su edad en años: "))
    ingreso_mensual = float(input("Ingrese su ingreso mensual (S/): "))

    # Validación de coherencia para evitar datos imposibles
    if edad < 0 or ingreso_mensual < 0:
        print("\n[ERROR]: No se permiten valores negativos.")
    else:
        # --- PROCESO (Evaluación con Operador Lógico 'and') ---
        # La rúbrica indica "mayor de 18 años" (edad > 18)
        if edad > 18 and ingreso_mensual >= 3000.0:
            debe_tributar = True
            # Cálculo del 5% del ingreso mensual
            impuesto_pagar = ingreso_mensual * 0.05
        else:
            debe_tributar = False
            impuesto_pagar = 0.0

        # --- SALIDA ESTRUCTURADA ---
        print("\n" + "=" * 45)
        print("         ESTADO DE SITUACIÓN TRIBUTARIA")
        print("=" * 45)
        print(f"Edad del contribuyente : {edad} años")
        print(f"Ingresos declarados    : S/ {ingreso_mensual:.2f}")
        print("-" * 45)

        if debe_tributar:
            print("STATUS             : [ X ] CONDICIÓN AFECTA")
            print("MENSAJE            : Usted está obligado a tributar.")
            print(f"MONTO A TRIBUTAR (5%): S/ {impuesto_pagar:.2f}")
        else:
            print("STATUS             : [   ] CONDICIÓN EXONERADA")
            print("MENSAJE            : Usted NO tiene que tributar.")
            print("MONTO A TRIBUTAR   : S/ 0.00")

        print("=" * 45)

except ValueError:
    print("\n" + "!" * 45)
    print("ERROR: Por favor, ingrese datos numéricos válidos.")
    print("!" * 45)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 4: Estímulo de Colegiatura (Estructura Condicional)
# =========================================================

print("--- SISTEMA DE MATRÍCULA Y BECAS ---")

try:
    # --- ENTRADAS ---
    num_materias = int(input("Ingrese el número de materias a cursar: "))
    promedio = float(input("Ingrese el promedio obtenido en el último periodo: "))

    # Validación de valores lógicos y límites de calificación estándar
    if num_materias <= 0 or promedio < 0 or promedio > 20:
        print("\n[ERROR]: Ingrese datos válidos (Materias > 0 y Promedio entre 0 y 20).")
    else:
        # --- PROCESO ---
        COSTO_MATERIA = 100.0
        costo_bruto = num_materias * COSTO_MATERIA

        # Evaluación de la condicional para aplicar el estímulo
        if promedio >= 18.0:
            # Descuento del 50% (0.50) sobre la mitad de los cursos matriculados
            descuento = (num_materias / 2 * COSTO_MATERIA) * 0.50
            aplica_beca = True
        else:
            descuento = 0.0
            aplica_beca = False

        # Cálculo del monto final neto
        total_pagar = costo_bruto - descuento

        # --- SALIDA ESTRUCTURADA ---
        print("\n" + "=" * 45)
        print("         LIQUIDACIÓN DE COLEGIATURA")
        print("=" * 45)
        print(f"Materias Inscritas   : {num_materias}")
        print(f"Promedio Registrado  : {promedio:.2f}")
        print("-" * 45)
        print(f"Costo Bruto Inicial  : S/ {costo_bruto:.2f}")

        if aplica_beca:
            print(f"Estímulo Académico   : [ X ] 50% desc. en mitad de cursos")
            print(f"Monto Descontado     : S/ {descuento:.2f}")
        else:
            print(f"Estímulo Académico   : [   ] No aplica (Promedio < 18)")
            print(f"Monto Descontado     : S/ 0.00")

        print("-" * 45)
        print(f"TOTAL NETO A PAGAR   : S/ {total_pagar:.2f}")
        print("=" * 45)

except ValueError:
    print("\n" + "!" * 45)
    print("ERROR: Por favor, ingrese únicamente valores numéricos.")
    print("!" * 45)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 5: Clasificación de Triángulos (Condicional Anidada)
# =========================================================

import math

print("--- SISTEMA DE GEOMETRÍA ANALÍTICA - TRIÁNGULOS ---")

try:
    # --- ENTRADAS ---
    l1 = float(input("Ingrese la longitud del lado 1: "))
    l2 = float(input("Ingrese la longitud del lado 2: "))
    l3 = float(input("Ingrese la longitud del lado 3: "))

    # --- VALIDACIÓN: Teorema de la desigualdad triangular ---
    # Un triángulo solo existe si la suma de dos de sus lados es mayor que el tercero.
    if (l1 + l2 <= l3) or (l1 + l3 <= l2) or (l2 + l3 <= l1) or (l1 <= 0 or l2 <= 0 or l3 <= 0):
        print("\n[ERROR]: Las medidas ingresadas no forman un triángulo válido.")
    else:
        # --- PROCESO Y SALIDA ESTRUCTURADA ---
        print("\n" + "=" * 45)
        print("         RESULTADO DEL ANÁLISIS GEOMÉTRICO")
        print("=" * 45)
        print(f"Lados ingresados: [{l1}], [{l2}], [{l3}]")
        print("-" * 45)

        # 1. Condición para Triángulo Equilátero
        if l1 == l2 == l3:
            perimetro = l1 + l2 + l3
            print("TIPO DE TRIÁNGULO : EQUILÁTERO (3 lados iguales)")
            print(f"CÁLCULO REQUERIDO : Perímetro = {perimetro:.2f}")

        # 2. Condición para Triángulo Escaleno
        elif l1 != l2 and l1 != l3 and l2 != l3:
            # Cálculo del área por Fórmula de Herón
            s = (l1 + l2 + l3) / 2
            area = math.sqrt(s * (s - l1) * (s - l2) * (s - l3))
            print("TIPO DE TRIÁNGULO : ESCALENO (3 lados diferentes)")
            print(f"CÁLCULO REQUERIDO : Área = {area:.2f} u²")

        # 3. Condición por descarte: Triángulo Isósceles
        else:
            print("TIPO DE TRIÁNGULO : ISÓSCELES (2 lados iguales)")
            # Identificamos cuáles son los lados iguales
            if l1 == l2:
                print(f"LADOS IGUALES     : Lado 1 y Lado 2 (Valor: {l1})")
            elif l1 == l3:
                print(f"LADOS IGUALES     : Lado 1 y Lado 3 (Valor: {l1})")
            else:
                print(f"LADOS IGUALES     : Lado 2 y Lado 3 (Valor: {l2})")

        print("=" * 45)

except ValueError:
    print("\n" + "!" * 45)
    print("ERROR: Ingrese únicamente dimensiones numéricas válidas.")
    print("!" * 45)
```

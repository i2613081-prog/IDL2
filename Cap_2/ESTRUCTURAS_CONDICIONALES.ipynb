```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 12: Mayor y Menor de Tres Números (Condicional Anidado)
# =========================================================

# --- ENTRADAS ---
num1 = float(input("Ingrese el primer número: "))
num2 = float(input("Ingrese el segundo número: "))
num3 = float(input("Ingrese el tercer número: "))

# --- PROCESO PARA HALLAR EL MAYOR ---
if num1 >= num2 and num1 >= num3:
    mayor = num1
elif num2 >= num1 and num2 >= num3:
    mayor = num2
else:
    mayor = num3

# --- PROCESO PARA HALLAR EL MENOR ---
if num1 <= num2 and num1 <= num3:
    menor = num1
elif num2 <= num1 and num2 <= num3:
    menor = num2
else:
    menor = num3

# --- SALIDA ESTRUCTURADA ---
print("\n" + "=" * 40)
print("         RESULTADO DE LA COMPARACIÓN")
print("=" * 40)
print(f"Números evaluados: [{num1}], [{num2}], [{num3}]")
print("-" * 40)
print(f"-> EL NÚMERO MAYOR ES : {mayor}")
print(f"-> EL NÚMERO MENOR ES : {menor}")
print("=" * 40)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 13: Control de Tributación (Condicional Compuesta)
# =========================================================

print("--- SISTEMA DE EVALUACIÓN SUNAT / TRIBUTARIA ---")

try:
    # --- ENTRADAS ---
    edad = int(input("Ingrese su edad en años: "))
    ingreso_mensual = float(input("Ingrese su ingreso mensual (S/): "))

    # Validación de coherencia para evitar datos imposibles
    if edad < 0 or ingreso_mensual < 0:
        print("\n[ERROR]: No se permiten valores negativos.")
    else:
        # --- PROCESO (Evaluación con Operador Lógico 'and') ---
        # La rúbrica indica "mayor de 18 años" (edad > 18)
        if edad > 18 and ingreso_mensual >= 3000.0:
            debe_tributar = True
            # Cálculo del 5% del ingreso mensual
            impuesto_pagar = ingreso_mensual * 0.05
        else:
            debe_tributar = False
            impuesto_pagar = 0.0

        # --- SALIDA ESTRUCTURADA ---
        print("\n" + "=" * 45)
        print("         ESTADO DE SITUACIÓN TRIBUTARIA")
        print("=" * 45)
        print(f"Edad del contribuyente : {edad} años")
        print(f"Ingresos declarados    : S/ {ingreso_mensual:.2f}")
        print("-" * 45)

        if debe_tributar:
            print("STATUS             : [ X ] CONDICIÓN AFECTA")
            print("MENSAJE            : Usted está obligado a tributar.")
            print(f"MONTO A TRIBUTAR (5%): S/ {impuesto_pagar:.2f}")
        else:
            print("STATUS             : [   ] CONDICIÓN EXONERADA")
            print("MENSAJE            : Usted NO tiene que tributar.")
            print("MONTO A TRIBUTAR   : S/ 0.00")

        print("=" * 45)

except ValueError:
    print("\n" + "!" * 45)
    print("ERROR: Por favor, ingrese datos numéricos válidos.")
    print("!" * 45)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 14: Estímulo de Colegiatura (Estructura Condicional)
# =========================================================

print("--- SISTEMA DE MATRÍCULA Y BECAS ---")

try:
    # --- ENTRADAS ---
    num_materias = int(input("Ingrese el número de materias a cursar: "))
    promedio = float(input("Ingrese el promedio obtenido en el último periodo: "))

    # Validación de valores lógicos y límites de calificación estándar
    if num_materias <= 0 or promedio < 0 or promedio > 20:
        print("\n[ERROR]: Ingrese datos válidos (Materias > 0 y Promedio entre 0 y 20).")
    else:
        # --- PROCESO ---
        COSTO_MATERIA = 100.0
        costo_bruto = num_materias * COSTO_MATERIA

        # Evaluación de la condicional para aplicar el estímulo
        if promedio >= 18.0:
            # Descuento del 50% (0.50) sobre la mitad de los cursos matriculados
            descuento = (num_materias / 2 * COSTO_MATERIA) * 0.50
            aplica_beca = True
        else:
            descuento = 0.0
            aplica_beca = False

        # Cálculo del monto final neto
        total_pagar = costo_bruto - descuento

        # --- SALIDA ESTRUCTURADA ---
        print("\n" + "=" * 45)
        print("         LIQUIDACIÓN DE COLEGIATURA")
        print("=" * 45)
        print(f"Materias Inscritas   : {num_materias}")
        print(f"Promedio Registrado  : {promedio:.2f}")
        print("-" * 45)
        print(f"Costo Bruto Inicial  : S/ {costo_bruto:.2f}")

        if aplica_beca:
            print(f"Estímulo Académico   : [ X ] 50% desc. en mitad de cursos")
            print(f"Monto Descontado     : S/ {descuento:.2f}")
        else:
            print(f"Estímulo Académico   : [   ] No aplica (Promedio < 18)")
            print(f"Monto Descontado     : S/ 0.00")

        print("-" * 45)
        print(f"TOTAL NETO A PAGAR   : S/ {total_pagar:.2f}")
        print("=" * 45)

except ValueError:
    print("\n" + "!" * 45)
    print("ERROR: Por favor, ingrese únicamente valores numéricos.")
    print("!" * 45)
```

```python
# =========================================================
# PROYECTO PRODUCTIVO IA - IDL 2
# Ejercicio 15: Clasificación de Triángulos (Condicional Anidada)
# =========================================================

import math

print("--- SISTEMA DE GEOMETRÍA ANALÍTICA - TRIÁNGULOS ---")

try:
    # --- ENTRADAS ---
    l1 = float(input("Ingrese la longitud del lado 1: "))
    l2 = float(input("Ingrese la longitud del lado 2: "))
    l3 = float(input("Ingrese la longitud del lado 3: "))

    # --- VALIDACIÓN: Teorema de la desigualdad triangular ---
    # Un triángulo solo existe si la suma de dos de sus lados es mayor que el tercero.
    if (l1 + l2 <= l3) or (l1 + l3 <= l2) or (l2 + l3 <= l1) or (l1 <= 0 or l2 <= 0 or l3 <= 0):
        print("\n[ERROR]: Las medidas ingresadas no forman un triángulo válido.")
    else:
        # --- PROCESO Y SALIDA ESTRUCTURADA ---
        print("\n" + "=" * 45)
        print("         RESULTADO DEL ANÁLISIS GEOMÉTRICO")
        print("=" * 45)
        print(f"Lados ingresados: [{l1}], [{l2}], [{l3}]")
        print("-" * 45)

        # 1. Condición para Triángulo Equilátero
        if l1 == l2 == l3:
            perimetro = l1 + l2 + l3
            print("TIPO DE TRIÁNGULO : EQUILÁTERO (3 lados iguales)")
            print(f"CÁLCULO REQUERIDO : Perímetro = {perimetro:.2f}")

        # 2. Condición para Triángulo Escaleno
        elif l1 != l2 and l1 != l3 and l2 != l3:
            # Cálculo del área por Fórmula de Herón
            s = (l1 + l2 + l3) / 2
            area = math.sqrt(s * (s - l1) * (s - l2) * (s - l3))
            print("TIPO DE TRIÁNGULO : ESCALENO (3 lados diferentes)")
            print(f"CÁLCULO REQUERIDO : Área = {area:.2f} u²")

        # 3. Condición por descarte: Triángulo Isósceles
        else:
            print("TIPO DE TRIÁNGULO : ISÓSCELES (2 lados iguales)")
            # Identificamos cuáles son los lados iguales
            if l1 == l2:
                print(f"LADOS IGUALES     : Lado 1 y Lado 2 (Valor: {l1})")
            elif l1 == l3:
                print(f"LADOS IGUALES     : Lado 1 y Lado 3 (Valor: {l1})")
            else:
                print(f"LADOS IGUALES     : Lado 2 y Lado 3 (Valor: {l2})")

        print("=" * 45)

except ValueError:
    print("\n" + "!" * 45)
    print("ERROR: Ingrese únicamente dimensiones numéricas válidas.")
    print("!" * 45)
```

```python
# ============================================================
#  EJERCICIO 16 - Hora N segundos despues
# ============================================================

print("="*50)
print("   RELOJ - HORA N SEGUNDOS DESPUES")
print("="*50)

# ---------- Leer la hora HH MM SS ----------
# Uso int(input()) para leer cada componente de la hora como numero entero
while True:
    try:
        hh = int(input("\nIngrese las horas (0-23): "))
        mm = int(input("Ingrese los minutos (0-59): "))
        ss = int(input("Ingrese los segundos (0-59): "))
        # Valido con operadores relacionales (>=, <=) y operador logico 'and'
        if 0 <= hh <= 23 and 0 <= mm <= 59 and 0 <= ss <= 59:
            break
        else:
            print("  Advertencia: ingrese valores validos de hora (HH 0-23, MM/SS 0-59).")
    except ValueError:
        print("  Advertencia: ingrese solo numeros enteros.")

# ---------- Leer N segundos a sumar ----------
while True:
    try:
        n_segundos = int(input("\nIngrese la cantidad de segundos a sumar: "))
        # Uso operador relacional >= para validar que no sea negativo
        if n_segundos >= 0:
            break
        else:
            print("  Advertencia: ingrese un numero positivo.")
    except ValueError:
        print("  Advertencia: ingrese un numero entero valido.")

# ============================================================
#  CALCULO
# ============================================================

# Convierto toda la hora a segundos totales (operadores aritmeticos * y +)
total_segundos = hh * 3600 + mm * 60 + ss

# Sumo los segundos indicados por el usuario
total_segundos = total_segundos + n_segundos

# Uso el operador modulo (%) para que el resultado no pase de 24 horas (86400 segundos)
total_segundos = total_segundos % 86400

# Recupero horas, minutos y segundos usando division entera (//) y modulo (%)
nueva_hh = total_segundos // 3600
resto = total_segundos % 3600
nueva_mm = resto // 60
nueva_ss = resto % 60

# ============================================================
#  RESULTADO
# ============================================================

print("\n" + "="*50)
print("   RESULTADO")
print("="*50)

print(f"\n  Hora ingresada      : {hh:02d}:{mm:02d}:{ss:02d}")
print(f"  Segundos a sumar    : {n_segundos}")
# :02d asegura que siempre se muestren 2 digitos (ej: 05 en lugar de 5)
print(f"  Hora resultante     : {nueva_hh:02d}:{nueva_mm:02d}:{nueva_ss:02d}")
print("="*50)
```

```python
# ============================================================
#  EJERCICIO 17 - Costo de llamada telefonica
# ============================================================

# Constantes con las tarifas (uso MAYUSCULAS por convencion para valores fijos)
TARIFA_1 = 1.00   # primeros 5 minutos
TARIFA_2 = 0.80   # siguientes 3 minutos
TARIFA_3 = 0.70   # siguientes 2 minutos
TARIFA_4 = 0.50   # a partir del minuto 11

print("="*50)
print("   CALCULO DE COSTO DE LLAMADA TELEFONICA")
print("="*50)

# ---------- Leer minutos de la llamada ----------
while True:
    try:
        minutos = int(input("\nIngrese la duracion de la llamada en minutos: "))
        if minutos > 0:
            break
        else:
            print("  Advertencia: la duracion debe ser mayor a 0.")
    except ValueError:
        print("  Advertencia: ingrese un numero entero valido.")

# ---------- Leer si es domingo ----------
# Uso .upper() para aceptar mayuscula o minuscula
while True:
    es_domingo = input("\n¿La llamada es un domingo? (S/N): ").strip().upper()
    if es_domingo in ("S", "N"):
        break
    print("  Advertencia: ingrese S para Si o N para No.")

# ---------- Leer si es turno nocturno (solo si es domingo) ----------
es_nocturno = "N"
if es_domingo == "S":
    while True:
        es_nocturno = input("¿Es turno nocturno? (S/N): ").strip().upper()
        if es_nocturno in ("S", "N"):
            break
        print("  Advertencia: ingrese S para Si o N para No.")

# ============================================================
#  CALCULO DEL COSTO POR TRAMOS
# ============================================================

costo = 0.0          # costo acumulado de la llamada
minutos_restantes = minutos

# Tramo 1: primeros 5 minutos
# Uso min() para no tomar mas minutos de los que realmente hay
tramo1 = min(minutos_restantes, 5)
costo += tramo1 * TARIFA_1
minutos_restantes -= tramo1

# Tramo 2: siguientes 3 minutos (minutos 6 a 8)
tramo2 = min(minutos_restantes, 3)
costo += tramo2 * TARIFA_2
minutos_restantes -= tramo2

# Tramo 3: siguientes 2 minutos (minutos 9 y 10)
tramo3 = min(minutos_restantes, 2)
costo += tramo3 * TARIFA_3
minutos_restantes -= tramo3

# Tramo 4: minuto 11 en adelante
tramo4 = minutos_restantes
costo += tramo4 * TARIFA_4

# ============================================================
#  APLICAR IMPUESTO
# ============================================================

impuesto = 0.0
if es_domingo == "S" and es_nocturno == "S":
    # Operador logico 'and': domingo Y nocturno => 5%
    impuesto = costo * 0.05
elif es_domingo == "S":
    # Solo domingo => 3%
    impuesto = costo * 0.03
# Si no es domingo, el impuesto queda en 0.0

total_pagar = costo + impuesto

# ============================================================
#  RESULTADOS
# ============================================================

print("\n" + "="*50)
print("   RESUMEN DE LA LLAMADA")
print("="*50)

print(f"\n  Duracion total            : {minutos} minutos")
print(f"  Minutos en tramo 1 (S/{TARIFA_1:.2f}) : {tramo1}")
print(f"  Minutos en tramo 2 (S/{TARIFA_2:.2f}) : {tramo2}")
print(f"  Minutos en tramo 3 (S/{TARIFA_3:.2f}) : {tramo3}")
print(f"  Minutos en tramo 4 (S/{TARIFA_4:.2f}) : {tramo4}")
print(f"\n  Costo sin impuesto        : S/ {costo:.2f}")
print(f"  Impuesto aplicado         : S/ {impuesto:.2f}")
print("  " + "-"*40)
print(f"  TOTAL A PAGAR             : S/ {total_pagar:.2f}")
print("="*50)
```

```python
# ============================================================
#  EJERCICIO 18 - Ordenar 3 numeros enteros (validar diferentes)
# ============================================================

print("="*50)
print("   ORDENAR 3 NUMEROS ENTEROS")
print("="*50)

# ---------- Leer 3 numeros diferentes ----------
while True:
    try:
        n1 = int(input("\nIngrese el primer numero: "))
        n2 = int(input("Ingrese el segundo numero: "))
        n3 = int(input("Ingrese el tercer numero: "))

        # Verifico que los 3 numeros sean diferentes entre si
        # Uso operador != (diferente de) y 'and' para combinar las 3 condiciones
        if n1 != n2 and n1 != n3 and n2 != n3:
            break
        else:
            print("  Advertencia: los 3 numeros deben ser diferentes entre si.")
    except ValueError:
        print("  Advertencia: ingrese numeros enteros validos.")

# ---------- Leer el orden deseado ----------
while True:
    orden = input("\n¿Como desea verlos? (A = Ascendente, D = Descendente): ").strip().upper()
    if orden in ("A", "D"):
        break
    print("  Advertencia: ingrese A para Ascendente o D para Descendente.")

# ============================================================
#  ORDENAMIENTO MANUAL (usando lista y sort())
# ============================================================

# Guardo los 3 numeros en una lista
numeros = [n1, n2, n3]

# sort() ordena la lista de menor a mayor (ascendente) de forma directa
numeros.sort()

# Si el usuario pidio descendente, invierto el orden de la lista con reverse()
if orden == "D":
    numeros.reverse()

# ============================================================
#  RESULTADOS
# ============================================================

print("\n" + "="*50)
print("   RESULTADO")
print("="*50)

print(f"\n  Numeros ingresados : {n1}, {n2}, {n3}")

if orden == "A":
    print("  Orden solicitado   : Ascendente")
else:
    print("  Orden solicitado   : Descendente")

# join convierte cada numero a texto para poder unirlos con una coma
print(f"  Numeros ordenados  : {', '.join(str(x) for x in numeros)}")
print("="*50)
```

```python
# ============================================================
#  EJERCICIO 19 - Compra de N camisas con descuento
# ============================================================

# Constante: precio fijo por camisa
PRECIO_CAMISA = 100   # S/ 100 por camisa

print("="*50)
print("   COMPRA DE CAMISAS CON DESCUENTO")
print("="*50)

# ---------- Leer cantidad de camisas ----------
while True:
    try:
        cantidad = int(input("\nIngrese la cantidad de camisas a comprar: "))
        if cantidad > 0:
            break
        else:
            print("  Advertencia: debe comprar al menos 1 camisa.")
    except ValueError:
        print("  Advertencia: ingrese un numero entero valido.")

# ============================================================
#  CALCULO DEL DESCUENTO SEGUN CANTIDAD
# ============================================================

# Monto sin descuento: precio fijo multiplicado por la cantidad
monto_sin_descuento = cantidad * PRECIO_CAMISA

# Estructura condicional para determinar el porcentaje de descuento
if 1 <= cantidad <= 4:
    porcentaje_descuento = 0.125   # 12.5%
elif 5 <= cantidad <= 8:
    porcentaje_descuento = 0.20    # 20%
else:
    # cantidad > 8
    porcentaje_descuento = 0.315   # 31.5%

# Calculo del monto de descuento y del total final
monto_descuento = monto_sin_descuento * porcentaje_descuento
monto_final = monto_sin_descuento - monto_descuento

# ============================================================
#  RESULTADOS
# ============================================================

print("\n" + "="*50)
print("   RESUMEN DE LA COMPRA")
print("="*50)

print(f"\n  Cantidad de camisas       : {cantidad}")
print(f"  Precio por camisa          : S/ {PRECIO_CAMISA:.2f}")
print(f"  Porcentaje de descuento    : {porcentaje_descuento*100:.1f}%")
print("\n  --- Detalle de la compra ---")
print(f"  Compra sin descuento       : S/ {monto_sin_descuento:.2f}")
print(f"  Monto del descuento        : S/ {monto_descuento:.2f}")
print("  " + "-"*40)
print(f"  TOTAL A PAGAR              : S/ {monto_final:.2f}")
print("="*50)
```

```python
# ============================================================
#  EJERCICIO 20 - Costo de llamada internacional por zona
# ============================================================

print("="*50)
print("   LLAMADAS INTERNACIONALES POR ZONA")
print("="*50)

# Muestro la tabla de zonas disponibles para que el usuario sepa que codigo ingresar
print("\n  CODIGO | ZONA GEOGRAFICA          | COSTO POR MINUTO (S/)")
print("  -------------------------------------------------------")
print("  10     | America del norte        | 2.2")
print("  12     | America del centro       | 2.5")
print("  20     | America del sur          | 1.2")
print("  22     | Asia                     | 3.5")
print("  30     | Europa                   | 3.0")
print("  32     | Africa                   | 3.2")

# ---------- Leer codigo de zona ----------
# Lista con los codigos validos para poder verificar con el operador 'in'
codigos_validos = [10, 12, 20, 22, 30, 32]

while True:
    try:
        codigo = int(input("\nIngrese el codigo de la zona: "))
        # 'in' verifica si el codigo esta dentro de la lista de codigos validos
        if codigo in codigos_validos:
            break
        else:
            print("  Advertencia: codigo no valido, revise la tabla.")
    except ValueError:
        print("  Advertencia: ingrese un numero entero valido.")

# ---------- Leer minutos hablados ----------
while True:
    try:
        minutos = float(input("Ingrese los minutos hablados: "))
        if minutos > 0:
            break
        else:
            print("  Advertencia: los minutos deben ser mayores a 0.")
    except ValueError:
        print("  Advertencia: ingrese un numero valido.")

# ============================================================
#  DETERMINAR ZONA Y COSTO SEGUN EL CODIGO
# ============================================================

# Estructura condicional encadenada (if / elif) para asociar codigo con zona y costo
if codigo == 10:
    zona = "America del norte"
    costo_minuto = 2.2
elif codigo == 12:
    zona = "America del centro"
    costo_minuto = 2.5
elif codigo == 20:
    zona = "America del sur"
    costo_minuto = 1.2
elif codigo == 22:
    zona = "Asia"
    costo_minuto = 3.5
elif codigo == 30:
    zona = "Europa"
    costo_minuto = 3.0
else:
    # Unico caso restante: codigo == 32
    zona = "Africa"
    costo_minuto = 3.2

# Calculo el costo total multiplicando minutos por el costo por minuto
costo_total = minutos * costo_minuto

# ============================================================
#  RESULTADOS
# ============================================================

print("\n" + "="*50)
print("   RESULTADO DE LA LLAMADA")
print("="*50)

print(f"\n  Codigo ingresado       : {codigo}")
print(f"  Zona geografica        : {zona}")
print(f"  Costo por minuto       : S/ {costo_minuto:.2f}")
print(f"  Minutos hablados       : {minutos}")
print("  " + "-"*40)
print(f"  COSTO TOTAL            : S/ {costo_total:.2f}")
print("="*50)
```

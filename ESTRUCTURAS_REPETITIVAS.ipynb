```python
# ============================================================
#  EJERCICIO 1 - Suma y promedio de pares e impares de 1 a N
# ============================================================

print("="*50)
print("   SUMA Y PROMEDIO DE PARES E IMPARES DE 1 A N")
print("="*50)

# ---------- Leer el valor de N ----------
while True:
    try:
        n = int(input("\nIngrese el valor de N: "))
        if n > 0:
            break
        else:
            print("  Advertencia: N debe ser mayor a 0.")
    except ValueError:
        print("  Advertencia: ingrese un numero entero valido.")

# ============================================================
#  CALCULO DE SUMAS Y CONTEOS
# ============================================================

# Inicializo en 0 las sumas y los contadores de pares e impares
suma_pares = 0
suma_impares = 0
cant_pares = 0
cant_impares = 0

# Recorro todos los numeros desde 1 hasta N (incluido) con un for
for numero in range(1, n + 1):
    # Operador modulo (%) para saber si el numero es par o impar
    if numero % 2 == 0:
        suma_pares += numero
        cant_pares += 1
    else:
        suma_impares += numero
        cant_impares += 1

# ============================================================
#  CALCULO DE PROMEDIOS
# ============================================================

# Uso operador ternario para evitar division entre cero si no hay pares o impares
promedio_pares = suma_pares / cant_pares if cant_pares > 0 else 0
promedio_impares = suma_impares / cant_impares if cant_impares > 0 else 0

# ============================================================
#  RESULTADOS
# ============================================================

print("\n" + "="*50)
print("   RESULTADOS")
print("="*50)

print(f"\n  N ingresado : {n}")

print("\n  --- Numeros pares ---")
print(f"  Cantidad de pares  : {cant_pares}")
print(f"  Suma de pares      : {suma_pares}")
print(f"  Promedio de pares  : {promedio_pares:.2f}")

print("\n  --- Numeros impares ---")
print(f"  Cantidad de impares : {cant_impares}")
print(f"  Suma de impares     : {suma_impares}")
print(f"  Promedio de impares : {promedio_impares:.2f}")
print("="*50)
```

```python
# ============================================================
#  EJERCICIO 2 - Clasificacion de 10 numeros
# ============================================================

CANTIDAD = 10   # cantidad fija de numeros a ingresar

print("="*50)
print("   CLASIFICACION DE 10 NUMEROS")
print("="*50)

# ---------- Leer los 10 numeros ----------
numeros = []   # lista vacia donde guardo los numeros ingresados

print(f"\nIngrese {CANTIDAD} numeros enteros:")
# range(1, CANTIDAD + 1) genera del 1 al 10 para numerar cada ingreso
for i in range(1, CANTIDAD + 1):
    while True:
        try:
            num = int(input(f"  Numero {i:>2}: "))
            numeros.append(num)   # append agrega el numero al final de la lista
            break
        except ValueError:
            print("    Advertencia: ingrese un numero entero valido.")

# ============================================================
#  CONTEO DE CADA CATEGORIA
# ============================================================

# Inicializo contadores en 0 para cada categoria
pares = 0
impares = 0
positivos = 0
neutros = 0
negativos = 0

# Recorro la lista con un for para clasificar cada numero
for num in numeros:
    # Operador modulo (%) para par/impar
    if num % 2 == 0:
        pares += 1
    else:
        impares += 1

    # Operadores relacionales para positivo, neutro (cero) o negativo
    if num > 0:
        positivos += 1
    elif num == 0:
        neutros += 1
    else:
        negativos += 1

# ============================================================
#  RESULTADOS
# ============================================================

print("\n" + "="*50)
print("   RESULTADOS DE LA CLASIFICACION")
print("="*50)

print(f"\n  Numeros ingresados : {numeros}")
print(f"\n  Cantidad de pares      : {pares}")
print(f"  Cantidad de impares    : {impares}")
print(f"  Cantidad de positivos  : {positivos}")
print(f"  Cantidad de neutros (0): {neutros}")
print(f"  Cantidad de negativos  : {negativos}")
print("="*50)
```

```python
# ============================================================
#  EJERCICIO 3 - Porcentajes respecto a un valor X
# ============================================================

print("="*50)
print("   PORCENTAJES RESPECTO A UN VALOR X")
print("="*50)

# ---------- Leer cantidad N de numeros ----------
while True:
    try:
        n = int(input("\nIngrese la cantidad de numeros (N): "))
        if n > 0:
            break
        else:
            print("  Advertencia: N debe ser mayor a 0.")
    except ValueError:
        print("  Advertencia: ingrese un numero entero valido.")

# ---------- Leer los N numeros ----------
numeros = []
print(f"\nIngrese los {n} numeros:")
for i in range(1, n + 1):
    while True:
        try:
            valor = float(input(f"  Numero {i:>2}: "))
            numeros.append(valor)
            break
        except ValueError:
            print("    Advertencia: ingrese un numero valido.")

# ---------- Leer el valor X ----------
while True:
    try:
        x = float(input("\nIngrese el valor de X: "))
        break
    except ValueError:
        print("  Advertencia: ingrese un numero valido.")

# ============================================================
#  CONTEO SEGUN CONDICIONES
# ============================================================

# Inicializo contadores para cada condicion
cant_menores = 0
cant_mayores = 0
cant_iguales = 0
cant_entre_1_y_x = 0
cant_ninguna = 0

# Recorro la lista de numeros para evaluar cada condicion
for valor in numeros:
    cumple_alguna = False   # bandera para saber si cumplio alguna condicion

    if valor < x:
        cant_menores += 1
        cumple_alguna = True

    if valor > x:
        cant_mayores += 1
        cumple_alguna = True

    if valor == x:
        cant_iguales += 1
        cumple_alguna = True

    # Operador logico 'and' para verificar el rango entre 1 y x
    if 1 <= valor <= x:
        cant_entre_1_y_x += 1
        cumple_alguna = True

    # Si no entro en ninguna condicion anterior, lo cuento aqui
    if not cumple_alguna:
        cant_ninguna += 1

# ============================================================
#  CALCULO DE PORCENTAJES
# ============================================================

# El porcentaje se calcula como (cantidad / total) * 100
pct_menores = (cant_menores / n) * 100
pct_mayores = (cant_mayores / n) * 100
pct_iguales = (cant_iguales / n) * 100
pct_entre = (cant_entre_1_y_x / n) * 100
pct_ninguna = (cant_ninguna / n) * 100

# ============================================================
#  RESULTADOS
# ============================================================

print("\n" + "="*50)
print("   RESULTADOS")
print("="*50)

print(f"\n Numeros ingresados : {numeros}")
print(f" Valor X               : {x}")

print("\n --- Porcentajes ---")
print(f"  Menores a X                  : {pct_menores:.1f}%")
print(f"  Mayores a X                  : {pct_mayores:.1f}%")
print(f"  Iguales a X                  : {pct_iguales:.1f}%")
print(f"  Comprendidos entre 1 y X     : {pct_entre:.1f}%")
print(f"  No cumplen ninguna condición : {pct_entre:.1f}%")
print("="*50)
```

```python
# ============================================================
#  EJERCICIO 4 - Maximo y minimo de N numeros
# ============================================================

print("="*50)
print("   MAXIMO Y MINIMO DE N NUMEROS")
print("="*50)

# ---------- Leer cantidad N de numeros ----------
while True:
    try:
        n = int(input("\nIngrese la cantidad de numeros (N): "))
        if n > 0:
            break
        else:
            print("  Advertencia: N debe ser mayor a 0.")
    except ValueError:
        print("  Advertencia: ingrese un numero entero valido.")

# ---------- Leer los N numeros ----------
numeros = []
print(f"\nIngrese los {n} numeros:")
for i in range(1, n + 1):
    while True:
        try:
            valor = float(input(f"  Numero {i:>2}: "))
            numeros.append(valor)
            break
        except ValueError:
            print("    Advertencia: ingrese un numero valido.")

# ============================================================
#  DETERMINAR MAXIMO Y MINIMO
# ============================================================

# max() y min() recorren la lista y devuelven el valor mas alto/bajo
numero_maximo = max(numeros)
numero_minimo = min(numeros)

# ============================================================
#  RESULTADOS
# ============================================================

print("\n" + "="*50)
print("   RESULTADOS")
print("="*50)

print(f"\n  Numeros ingresados : {numeros}")
print(f"\n  Valor MAXIMO : {numero_maximo}")
print(f"  Valor MINIMO : {numero_minimo}")
print("="*50)
```

```python
# ============================================================
#  EJERCICIO 5 - Ahorro mensual durante un año
# ============================================================

MESES = 12   # cantidad fija de meses en un año

# Lista con los nombres de los meses para mostrar mensajes mas claros
nombres_meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
                  "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"]

print("="*50)
print("   AHORRO MENSUAL DURANTE UN AÑO")
print("="*50)

ahorro_total = 0.0   # acumulador del ahorro total

print("\nIngrese el monto que ahorrara cada mes:")
# for con range(MESES) recorre de 0 a 11, que uso para indexar nombres_meses
for i in range(MESES):
    while True:
        try:
            deposito = float(input(f"  {nombres_meses[i]}: S/ "))
            if deposito >= 0:
                break
            else:
                print("    Advertencia: el monto no puede ser negativo.")
        except ValueError:
            print("    Advertencia: ingrese un numero valido.")

    # Acumulo el deposito del mes al ahorro total
    ahorro_total += deposito

    # Muestro cuanto se lleva ahorrado hasta el momento
    print(f"    -> Ahorro acumulado hasta {nombres_meses[i]}: S/ {ahorro_total:.2f}")

# ============================================================
#  RESULTADO FINAL
# ============================================================

print("\n" + "="*50)
print("   RESULTADO FINAL")
print("="*50)
print(f"\n  Ahorro total despues de {MESES} meses: S/ {ahorro_total:.2f}")
print("="*50)
```

```python
# ============================================================
#  EJERCICIO 6 - Cronometro (hh:mm:ss) referencial
# ============================================================

print("="*50)
print("   CRONOMETRO REFERENCIAL (hh:mm:ss)")
print("="*50)

# ---------- Leer cuantos segundos debe avanzar el cronometro ----------
while True:
    try:
        duracion = int(input("\nIngrese cuantos segundos debe avanzar el cronometro: "))
        if duracion > 0:
            break
        else:
            print("  Advertencia: la duracion debe ser mayor a 0.")
    except ValueError:
        print("  Advertencia: ingrese un numero entero valido.")

# ============================================================
#  SIMULACION DEL CRONOMETRO
# ============================================================

print("\nIniciando cronometro...\n")

segundo_actual = 0   # contador de segundos transcurridos

# Uso while para avanzar segundo por segundo hasta llegar a la duracion indicada
while segundo_actual <= duracion:
    # Division entera (//) y modulo (%) para obtener horas, minutos y segundos
    hh = segundo_actual // 3600
    mm = (segundo_actual % 3600) // 60
    ss = segundo_actual % 60

    # :02d muestra siempre 2 digitos (ej: 05 en vez de 5)
    print(f"  {hh:02d}:{mm:02d}:{ss:02d}")

    segundo_actual += 1   # avanzo un segundo

print("\n" + "="*50)
print("   CRONOMETRO FINALIZADO")
print("="*50)
```

```python
# ============================================================
#  EJERCICIO 7 - Factorial de dos numeros
# ============================================================

print("="*50)
print("   FACTORIAL DE DOS NUMEROS")
print("="*50)

# ---------- Leer los dos numeros ----------
while True:
    try:
        a = int(input("\nIngrese el primer numero: "))
        if a >= 0:
            break
        else:
            print("  Advertencia: el numero debe ser mayor o igual a 0.")
    except ValueError:
        print("  Advertencia: ingrese un numero entero valido.")

while True:
    try:
        b = int(input("Ingrese el segundo numero: "))
        if b >= 0:
            break
        else:
            print("  Advertencia: el numero debe ser mayor o igual a 0.")
    except ValueError:
        print("  Advertencia: ingrese un numero entero valido.")

# ============================================================
#  CALCULO DEL FACTORIAL DE CADA NUMERO
# ============================================================

# Calculo el factorial de 'a' multiplicando desde 1 hasta a
factorial_a = 1
for i in range(1, a + 1):
    factorial_a = factorial_a * i

# Calculo el factorial de 'b' multiplicando desde 1 hasta b
factorial_b = 1
for i in range(1, b + 1):
    factorial_b = factorial_b * i

# ============================================================
#  DETERMINAR HASTA QUE VALOR SON IGUALES LAS MULTIPLICACIONES
# ============================================================

# El menor de los dos numeros define hasta donde se pueden comparar
menor = min(a, b)

# Recorro desde 1 hasta el menor de los dos numeros comparando
# las multiplicaciones parciales (productos acumulados)
producto_a = 1
producto_b = 1
ultimo_igual = 0   # guarda el ultimo multiplo donde ambos productos coinciden

for i in range(1, menor + 1):
    producto_a = producto_a * i
    producto_b = producto_b * i
    # Como ambos parten igual y se multiplican por el mismo i, siempre coinciden
    # hasta el valor 'menor'; por eso el ultimo igual es el propio 'menor'
    if producto_a == producto_b:
        ultimo_igual = i

# ============================================================
#  RESULTADOS
# ============================================================

print("\n" + "="*50)
print("   RESULTADOS")
print("="*50)

print(f"\n  {a}! = {factorial_a}")
print(f"  {b}! = {factorial_b}")
print(f"\n  Ambos factoriales son iguales en sus multiplicaciones hasta el multiplo {ultimo_igual}")
print("="*50)
```

```python
# ============================================================
#  EJERCICIO 8 - Registro de Asistentes a una Fiesta
# ============================================================

# Listas para guardar los datos de cada asistente
# cada indice i corresponde a los datos de la misma persona
nombres = []
edades = []
sexos = []

print("="*50)
print("   REGISTRO DE ASISTENTES A LA FIESTA")
print("="*50)

# Variable bandera para controlar el bucle principal
continuar = True

# Bucle principal: se repite mientras se quieran registrar mas personas
while continuar:

    # ------- Nombre -------
    nombre = input("\nIngrese el nombre del asistente: ").strip()
    while nombre == "":
        print("  Advertencia: el nombre no puede estar vacio.")
        nombre = input("Ingrese el nombre del asistente: ").strip()

    # ------- Edad -------
    while True:
        try:
            edad = int(input(f"Ingrese la edad de {nombre}: "))
            if edad <= 0:
                print("  Advertencia: la edad debe ser un numero positivo.")
            else:
                break
        except ValueError:
            print("  Advertencia: ingrese un numero entero valido.")

    # ------- Sexo -------
    while True:
        sexo = input(f"Ingrese el sexo de {nombre} (M/F): ").strip().upper()
        if sexo in ("M", "F"):
            break
        print("  Advertencia: ingrese solo M (Masculino) o F (Femenino).")

    # ------- Guardar datos -------
    nombres.append(nombre)
    edades.append(edad)
    sexos.append(sexo)
    print(f"  {nombre} registrado/a correctamente.")

    # ------- ¿Continuar? -------
    while True:
        respuesta = input("\n¿Desea registrar a otra persona? (S/N): ").strip().upper()
        if respuesta in ("S", "N"):
            break
        print("  Advertencia: ingrese S para Si o N para No.")

    continuar = (respuesta == "S")

print("\n" + "="*50)
print("   Registro finalizado.")
print("="*50)

# ============================================================
#  CALCULOS Y RESULTADOS
# ============================================================

# len() devuelve la cantidad de elementos, equivale al total de asistentes
total = len(nombres)

# List comprehension: recorro los indices y selecciono las edades
# cuyo sexo correspondiente sea 'M' o 'F'
edades_hombres = [edades[i] for i in range(total) if sexos[i] == "M"]
edades_mujeres = [edades[i] for i in range(total) if sexos[i] == "F"]

cant_hombres = len(edades_hombres)
cant_mujeres = len(edades_mujeres)

# Operador ternario para evitar division entre cero si no hubo hombres/mujeres
prom_hombres = sum(edades_hombres) / cant_hombres if cant_hombres > 0 else 0
prom_mujeres = sum(edades_mujeres) / cant_mujeres if cant_mujeres > 0 else 0

# min() y max() devuelven la edad menor y mayor de toda la lista
edad_min = min(edades)
edad_max = max(edades)

# Busco los nombres de quienes tienen la edad minima o maxima (puede haber varios)
personas_min = [nombres[i] for i in range(total) if edades[i] == edad_min]
personas_max = [nombres[i] for i in range(total) if edades[i] == edad_max]

print("\n" + "="*50)
print("   RESULTADOS DE LA FIESTA")
print("="*50)

print(f"\n  Total de asistentes : {total}")
print(f"  Hombres             : {cant_hombres}")
print(f"  Mujeres             : {cant_mujeres}")

print("\n  --- Promedio de edades por sexo ---")
if cant_hombres > 0:
    print(f"  Promedio edad hombres : {prom_hombres:.2f} años")
else:
    print("  No hubo hombres registrados.")
if cant_mujeres > 0:
    print(f"  Promedio edad mujeres : {prom_mujeres:.2f} años")
else:
    print("  No hubo mujeres registradas.")

print("\n  --- Edades extremas ---")
print(f"  Menor edad : {edad_min} años -> {', '.join(personas_min)}")
print(f"  Mayor edad : {edad_max} años -> {', '.join(personas_max)}")

print("\n  --- Lista de asistentes ---")
print(f"  {'#':<4} {'Nombre':<20} {'Edad':>5}  {'Sexo':<8}")
print("  " + "-"*42)
for i in range(total):
    sexo_label = "Hombre" if sexos[i] == "M" else "Mujer"
    print(f"  {i+1:<4} {nombres[i]:<20} {edades[i]:>5}  {sexo_label:<8}")
print("="*50)
```

```python
# ============================================================
#  EJERCICIO 9 - Producto 100 x 98 x 96 x ... x 1
# ============================================================

print("="*50)
print("   PRODUCTO 100 x 98 x 96 x ... x 1")
print("="*50)

# Lista vacia donde voy a guardar los numeros de la secuencia
numeros = []

# range(100, 0, -2) genera numeros pares desde 100 hasta 2 (descendente de 2 en 2)
for valor in range(100, 0, -2):
    numeros.append(valor)

# Al final agrego manualmente el numero 1, segun pide el ejercicio
numeros.append(1)

# ============================================================
#  CALCULO DEL PRODUCTO
# ============================================================

# Inicio el producto en 1 (elemento neutro de la multiplicacion)
producto = 1

# Recorro la lista con un for y voy multiplicando cada valor
for valor in numeros:
    producto = producto * valor

# ============================================================
#  RESULTADOS
# ============================================================

print("\n" + "="*50)
print("   RESULTADOS")
print("="*50)

# Muestro la secuencia uniendo cada numero con " x " usando join
secuencia_texto = " x ".join(str(n) for n in numeros)
print(f"\n  Secuencia : {secuencia_texto}")
print(f"\n  PRODUCTO TOTAL = {producto}")
print("="*50)
```

```python
# ============================================================
#  EJERCICIO 10 - Notas de Estudiantes
# ============================================================

# Constantes del programa
NOTA_MIN_APROBATORIA = 13   # Nota minima para aprobar el curso
N_MIN = 5                   # Cantidad minima de estudiantes permitida
N_MAX = 20                  # Cantidad maxima de estudiantes permitida

print("="*50)
print("   REGISTRO DE NOTAS DE ESTUDIANTES")
print("="*50)

# ---------- Leer cantidad de estudiantes ----------
while True:
    try:
        n = int(input(f"\nIngrese la cantidad de estudiantes ({N_MIN}-{N_MAX}): "))
        if N_MIN <= n <= N_MAX:
            break
        print(f"  Advertencia: debe ingresar un valor entre {N_MIN} y {N_MAX}.")
    except ValueError:
        print("  Advertencia: ingrese un numero entero valido.")

# ---------- Leer notas ----------
notas = []

print(f"\nIngrese las notas de los {n} estudiantes (0 a 20):")
print("-"*40)

for i in range(1, n + 1):
    while True:
        try:
            nota = int(input(f"  Estudiante {i:>2}: "))
            if 0 <= nota <= 20:
                notas.append(nota)
                break
            else:
                print("    Advertencia: la nota debe estar entre 0 y 20.")
        except ValueError:
            print("    Advertencia: ingrese un numero entero valido.")

print("\n  Todas las notas registradas.")
print("="*50)

# ============================================================
#  CALCULOS
# ============================================================

total_alumnos   = len(notas)

nota_mayor      = max(notas)
nota_menor      = min(notas)

cant_mayor      = notas.count(nota_mayor)
cant_menor      = notas.count(nota_menor)

promedio        = sum(notas) / total_alumnos

cant_pares      = sum(1 for n in notas if n % 2 == 0)
cant_impares    = sum(1 for n in notas if n % 2 != 0)

aprobados       = sum(1 for n in notas if n >= NOTA_MIN_APROBATORIA)
desaprobados    = total_alumnos - aprobados

pct_aprobados   = (aprobados   / total_alumnos) * 100
pct_desaprobados= (desaprobados / total_alumnos) * 100

# ============================================================
#  MOSTRAR RESULTADOS
# ============================================================

print("\n" + "="*50)
print("   RESULTADOS DEL SALON")
print("="*50)

print(f"\n  Total de estudiantes           : {total_alumnos}")
print(f"  Nota aprobatoria minima        : {NOTA_MIN_APROBATORIA}")

print("\n  --- Notas extremas ---")
print(f"  a) Mayor nota                  : {nota_mayor}")
print(f"  b) Menor nota                  : {nota_menor}")
print(f"  c) Alumnos con mayor nota      : {cant_mayor}")
print(f"  d) Alumnos con menor nota      : {cant_menor}")

print("\n  --- Promedios y distribucion ---")
print(f"  e) Promedio del salon          : {promedio:.2f}")
print(f"  f) Notas pares                 : {cant_pares}")
print(f"  g) Notas impares               : {cant_impares}")

print("\n  --- Aprobados / Desaprobados ---")
print(f"  h) Alumnos aprobados           : {aprobados}")
print(f"  i) Alumnos desaprobados        : {desaprobados}")
print(f"  j) % Aprobados                 : {pct_aprobados:.1f}%")
print(f"  k) % Desaprobados              : {pct_desaprobados:.1f}%")

print("\n  --- Detalle de notas ---")
print(f"  {'#':<5} {'Nota':>5}  {'Estado':<14} {'Par/Impar'}")
print("  " + "-"*38)
for idx, nota in enumerate(notas, 1):
    estado    = "Aprobado"    if nota >= NOTA_MIN_APROBATORIA else "Desaprobado"
    paridad   = "Par"         if nota % 2 == 0               else "Impar"
    print(f"  {idx:<5} {nota:>5}  {estado:<14} {paridad}")

print("="*50)
```
